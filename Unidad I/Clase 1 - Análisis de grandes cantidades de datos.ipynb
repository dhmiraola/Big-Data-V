## Variables dummies

```python
subset.dtypes 
```

```python
subset['sexo_num'] = pd.get_dummies(subset['sexo'], drop_first=True)
subset.head()
```

```python
subset['sexo_num'].dtypes
```

```python
# Transformar la columna bool o booleana a integer
subset['sexo_num'] = pd.get_dummies(subset['sexo_num'], drop_first=True).astype(int)
subset['sexo_num'].dtypes
```

```python
# Crear una funcion que separe las edades en rangos

def rango_edad(edad):
    if edad <= 18:
        return 'Menor de 18'
    elif 19 <= edad <= 35:
        return '19-35'
    elif 36 <= edad <= 50:
        return '36-50'
    elif 51 <= edad <= 65:
        return '51-65'
    else:
        return '+65'
```

```python
# Aplicar la funcion a la columna edad y creamos una columna con los nuevos datos
subset['rango_edad'] = subset['edad'].apply(rango_edad)
subset.head()
```

## Outliers o valores atipicos

```python
# Identificar outliers mediante el metodo 'Recorrido intercuartil'
Q1 = subset['edad'].quantile(0.25) # 1er cuartil
Q3 = subset['edad'].quantile(0.75) # 3er cuartil

IQR = Q3 - Q1

outliers = subset[(subset['edad']  < (Q1 - 1.5 * IQR)) | (subset['edad'] > (Q3 + 1.5 * IQR))]
outliers
```

```python
# Identificamos outliers mediante un boxplot
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.boxplot(subset['edad'].dropna()) # el dropna sirve para eliminar las columnas NaN
plt.title('Boxplot de Edad')
plt.show()
```

```python
# borrar el outlier
data_cleaned = subset[subset['edad'] < 120 ] # asumiendo el valor del outlier

plt.figure(figsize=(10,6))
plt.boxplot(data_cleaned['edad'].dropna()) # el dropna sirve para eliminar las columnas NaN
plt.title('Boxplot de Edad')
plt.show()
```

```python
# imputar mediante el promedio
promedio_edad = subset['edad'].mean()
subset['edad'] = subset['edad'].apply(lambda x: promedio_edad if x == 120 else x)
subset['edad'].describe()
```
